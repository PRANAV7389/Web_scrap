```python
import pandas as pd
```

```python
web1 = "https://www.screener.in/screens/599148/companies-by-market-cap/?page=1"
table = pd.read_html(web1)
len(table1)
```

```python
table1 = table1[0]
```

```python
table1.head()
```

```python
type(table1)
```

```python

for i in range(2,30):
  web = f"https://www.screener.in/screens/599148/companies-by-market-cap/?page={i}"
  table = pd.read_html(web)[0]
  combined_df = pd.concat([table1,table], axis=0)
```

```python
combined_df.head()
```

```python
combined_df.shape
```

```python
combined_df.describe()
```

```python
combined_df = combined_df[combined_df["S.No."]!="S.No."]
```

```python
combined_df.describe()
```

```python
combined_df.columns
```

```python
combined_df.dropna(inplace=True)
```

```python
combined_df
```

```python
companies_by_qtr_profit = combined_df.loc[0:,["Name", "NP Qtr  Rs.Cr."]]
```

```python
companies_by_qtr_profit['NP Qtr  Rs.Cr.'] =companies_by_qtr_profit['NP Qtr  Rs.Cr.'].astype(float)
```

```python
companies_by_qtr_profit=companies_by_qtr_profit.sort_values("NP Qtr  Rs.Cr.",ascending=False)
```

```python
from matplotlib import pyplot as plt
companies_by_qtr_profit['NP Qtr  Rs.Cr.'].plot(kind='line', figsize=(8, 4), title='NP Qtr  Rs.Cr.')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
plt.figure(figsize=(20,10))
plt.barh(companies_by_qtr_profit['Name'][0:10],companies_by_qtr_profit['NP Qtr  Rs.Cr.'][0:10],color="darkblue")
plt.xlabel('Net Profit (Rs. Cr.)')
plt.ylabel('Company')
plt.title('Top 10 Companies by Net Profit')
plt.show()
```
